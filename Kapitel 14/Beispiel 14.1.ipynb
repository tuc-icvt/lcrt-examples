```python
### Import
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

In diesem Beispiel wird ein realer Reaktor durch Aufgabe einer Sprungfunktion am Eintritt untersucht. 


Es folgt die Parametrierung der Messwerte gemäß der Tabelle 14.1.

```python
Theta_M     = np.array((0, 0.262, 0.524, 0.786, 1.048, 1.310, 1.572, 1.834, 2.096, 2.620)) # dimensionslose Verweilzeit in 1
Messwerte   = np.array((0, 0.034, 0.221, 0.485, 0.706, 0.849, 0.928, 0.968, 0.986, 0.998)) # relative Tracerkonzentration am Reaktoraustritt in 1
```

Im Folgenden sollen die Messwerte mit der Verweilzeitsummenfunktion (Gleichung 14.44) beschrieben werden:

\begin{align*}
F(\Theta) = 1 - \exp\left( - K\, \Theta \right) \sum^K_{k=1} \frac{(K\, \Theta)^{k-1}}{(k-1)!}
\end{align*}

Dabei sollen zwei verschiedene Kesselzahlen $\left(K = 3 \text{ und } K = 4 \right)$ getestet werden.
Es folgt die Implementierung der Verweilzeitsummenfunktion _VZ_.

```python
def VZ(Theta,K):
    '''
    Verweilzeitsummenfunktion

    Parameter:
    ----------
    Theta   : dimensionslose Zeit
    K       : Anzahl der Kaskaden in 1
    k_vek   : Vektor der Größe K mit Einträgen 1,2, ...,K 
    sumvek  : Liste zur Bildung der Summe in Gleichung 14.1
    F       : Verweilzeitsumme in 1

    '''
    k_vek  = np.arange(1,K+1)  # Vektor der Größe K mit Einträgen 1,2, ...,K  
    sumvek = []                # Leere Liste 

    # Berechnung der Summe
    for k in k_vek:
        # Es werden alle Summanden separat als Funktion von k bestimmt
        fakt = factorial(int(k-1))                  # Faktorisieren von k
        sumvek.append((K * Theta)**(k-1) / fakt)    # berechnen eines Summanden der Summe und Speichern des Ergebnisses in der Liste
    sum_ = sum(sumvek)                              # Bilden der Summe
    F = 1- np.exp(- K * Theta) * sum_               # Berechnen der Verweilzeitsumme
    return F
```

Die Verweilzeitsummen werden nun für drei und vier Kessel berechnet.

```python
Theta_plot = np.linspace(0,3,100)   # Definieren eines dimensionslosen Zeitintervalls 

F_K3_plot = VZ(Theta_plot,3)        # Verweilzeitsummenfunktion für drei Kessel
F_K4_plot = VZ(Theta_plot,4)        # Verweilzeitsummenfunktion für vier Kessel
```

Abschließend werden die Verweilzeitsummen mit den Messwerten über der dimensionslosen Verweilzeit aufgetragen.

```python
fig, ax = plt.subplots(1, 1, figsize = (5, 5))
ax.plot(Theta_plot, F_K3_plot, label = r'$K=3$', linestyle = '-' , color = 'black')
ax.plot(Theta_plot, F_K4_plot, label = r'$K=4$', linestyle = '--', color = 'black')
ax.scatter(Theta_M, Messwerte, label = 'Messwerte', color = 'black', clip_on = False)
ax.set_xlim(0,3)
ax.set_ylim(-0.05,1.05)
ax.tick_params(axis="y",direction="in", right = True)
ax.tick_params(axis="x",direction="in", top = True)
ax.set_xticks(np.arange(0, 3.1, step = 1))
ax.set_yticks(np.arange(0, 1.1, step = 0.2))
ax.set_ylabel(r'$F\,/\,1$')
ax.set_xlabel(r'$\Theta \,/\,1$')
ax.legend(frameon = False, alignment = 'left')
plt.tight_layout()
plt.show()
```
